```python
simdi = datetime.datetime.now()

milyonun_6_da_biri = 1000000 / 6

if simdi.microsecond < milyonun_6_da_biri:
  print("Bir")
elif simdi.microsecond < milyonun_6_da_biri * 2:
  print("İki")
elif simdi.microsecond < milyonun_6_da_biri * 3:
  print("Üç")
elif simdi.microsecond < milyonun_6_da_biri * 4:
  print("Dört")
elif simdi.microsecond < milyonun_6_da_biri * 5:
  print("Beş")
else:
  print("Altı")
```

```python
simdi = datetime.datetime.now()

milyonun_6_da_biri = 1000000 / 6
sayac = 0

while not (simdi.microsecond < milyonun_6_da_biri * 5 and simdi.microsecond > milyonun_6_da_biri * 4):
  sayac = sayac + 1
  simdi = datetime.datetime.now()
  print(sayac, simdi.microsecond)


print(sayac)
```

```python
def a_ve_b_yi_bul(x, y):
  a = sum((x - x.mean()) * (y - y.mean())) / sum((x - x.mean()) ** 2)
  b = y.mean() - a * x.mean()
  return a, b
```
